## 1.Data Understanding and Exploration

```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('BoomBikes.csv')
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
#Assigning string values to different seasons instead of numeric values

# 1 - Spring
dataset.loc[(dataset['season'] == 1), 'season'] = 'spring'

# 2 - Summer
dataset.loc[(dataset['season'] == 2),'season'] = 'summer'

# 3 - Fall
dataset.loc[(dataset['season'] == 3),'season'] = 'fall'

# 4 - Winter
dataset.loc[(dataset['season'] == 4),'season'] = 'winter'

```

```python
dataset['season'].astype('category').value_counts()
```

```python
# 0=2018   1=2019 Year

dataset['yr'].astype('category').value_counts()
```

```python
#Asssigning string  values to different months having numeric values
def object_map_mnths(x):
    return x.map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'})
```

```python
dataset[['mnth']]=dataset[['mnth']].apply(object_map_mnths)
```

```python
dataset['mnth'].astype('category').value_counts()
```

```python
dataset['holiday'].astype('category').value_counts()
```

```python
def object_map_weekday(x):
    return x.map({1:'Mon',2:'Tues',3:'Wed',4:'Thurs',5:'Fri',6:'Sat',0:'Sun'})
```

```python
dataset[['weekday']]=dataset[['weekday']].apply(object_map_weekday)
```

```python
# 1 = Clear, few clouds, Partly Cloudy
dataset.loc[(dataset['weathersit'] == 1), 'weathersit'] = 'A'

# 2 = Mist, Cloudy
dataset.loc[(dataset['weathersit'] == 2), 'weathersit'] ='B'

# 3 = Light Snow, Heavy Rain
dataset.loc[(dataset['weathersit'] == 3), 'weathersit'] ='C'
```

```python
dataset['weathersit'].astype('category').value_counts()
```

## 3.Data Preparation

```python
dataset_categorical=dataset.select_dtypes(include=['object'])
```

```python
dataset_categorical.head()
```

```python
dataset_dummies = pd.get_dummies(dataset_categorical,drop_first=True)
dataset_dummies.head()
```

```python
#Drop categorical variable columns
dataset = dataset.drop(list(dataset_categorical.columns), axis=1)
dataset
```

```python
#Concatenating dummy variables with the dataset
dataset=pd.concat([dataset,dataset_dummies],axis=1)
```

```python
dataset.head()
```

```python
dataset = dataset.drop(['instant','dteday'], axis=1,inplace=False)
dataset.head()
```

## 4.Model Building and Evaluation

```python
#importing libraries
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
#Split the dataset into train and test datasets
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test=train_test_split(dataset,train_size=0.7,test_size=0.3,random_state=42)
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
# Apply scaling to all variables except dummy variables
var = ['temp','hum','windspeed','casual','registered','cnt']
df_train[var]=scaler.fit_transform(df_train[var])
```

```python
df_train.describe()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df_train.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
#Dividing into X and Y
Y_train = df_train.pop('cnt')
X_train = df_train.drop(['casual','registered'],axis=1)
```

```python
X_train.head()
```

```python
#np.asarray(df_train)
```

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)
lr = sm.OLS(Y_train,X_train_lm).fit()
```

```python
lr.params
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
lr.summary()
```

```python
#importing rfe
from sklearn.feature_selection import RFE
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
#15 features to select
rfe1 = RFE(estimator=lm, n_features_to_select=15)
rfe1.fit(X_train, Y_train)

print(rfe1.support_)
print(rfe1.ranking_)
```

```python
col1= X_train.columns[rfe1.support_]
```

```python
col1
```

```python
X_train_rfe1=X_train[col1]
X_train_rfe1= sm.add_constant(X_train_rfe1)
lm1 = sm.OLS(Y_train,X_train_rfe1).fit()
lm1.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
a = X_train_rfe1.drop('const',axis=1)
```

```python
#Evaluating Variance Inflation Factors
vif = pd.DataFrame()
vif['features']=a.columns
vif['VIF'] = [variance_inflation_factor(a.values,i)for i in range(a.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF", ascending =False)
vif
```

```python
lm2 = LinearRegression()
#7 features to select
rfe2 = RFE(estimator=lm2, n_features_to_select=7)
rfe2.fit(X_train, Y_train)

print(rfe2.support_)
print(rfe2.ranking_)
```

```python
col2= X_train.columns[rfe2.support_]

X_train_rfe2=X_train[col1]
X_train_rfe2= sm.add_constant(X_train_rfe2)
lm2 = sm.OLS(Y_train,X_train_rfe2).fit()
lm2.summary()
```

```python
b = X_train_rfe2.drop('const',axis=1)

vif1 = pd.DataFrame()
vif1['features']=b.columns
vif1['VIF'] = [variance_inflation_factor(b.values,i)for i in range(b.shape[1])]
vif1['VIF']=round(vif1['VIF'],2)
vif1 = vif1.sort_values(by = "VIF", ascending =False)
vif1
```

```python
Y_train_cnt = lm2.predict(X_train_rfe2)
```

```python
fig = plt.figure()
sns.distplot((Y_train,Y_train_cnt),bins=20)
```
